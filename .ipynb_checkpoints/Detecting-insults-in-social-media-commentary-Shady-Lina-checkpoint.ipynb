```python


nltk.download('stopwords')
nltk.download('punkt')

def remove_symbols_and_numbers(text):
    # Define a string containing all symbols and numbers you want to remove
    symbols_and_numbers = string.punctuation + string.digits

    # Use str.translate to remove symbols and numbers from the text
    translator = str.maketrans('', '', symbols_and_numbers)
    clean_text = text.translate(translator)

    return clean_text

def remove_stop_words(text):
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text)
    filtered_tokens = [
        token for token in tokens if token.lower() not in stop_words]
    return " ".join(filtered_tokens)

# Assuming df is your DataFrame with a 'comment' column
df['cleaned_comment'] = df['Comment'].apply(remove_symbols_and_numbers)
df['cleaned_comment'] = df['cleaned_comment'].apply(remove_stop_words)
```

### 5.1 logistic regression 
Logistic regression is a widely used statistical and machine learning model that is particularly suited for binary classification tasks, making it a valuable tool for insult detection.


##### hyperparameters for Logistic Regression:

. C (Regularization Strength): Controls the strength of regularization. Smaller values increase regularization, preventing overfitting; larger values reduce regularization, potentially leading to overfitting.

. Penalty (L1 or L2): Chooses the type of regularization. 'L1' for Lasso and 'L2' for Ridge. Affects the regularization term in the loss function.

. Solver: Optimization algorithm for model fitting. Common options include 'liblinear', 'newton-cg', 'lbfgs', 'sag', and 'saga'. Influences training time and convergence.

. Max Iterations (max_iter): Sets the maximum number of iterations for convergence. If the solver doesn't converge within this limit, it may not find the optimal solution.







```python
tfidf_vectorizer = TfidfVectorizer()
X = tfidf_vectorizer.fit_transform(df['tokenized-comment'].apply(lambda x: ' '.join(x)))

# Initialize the logistic regression model
logistic_reg = LogisticRegression(C=1.0, penalty='l1', solver='liblinear', max_iter=100)

# Train the model on the training data
logistic_reg.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = logistic_reg.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)

# Print the evaluation metrics
print(f"Accuracy: {accuracy:.2f}")
```
